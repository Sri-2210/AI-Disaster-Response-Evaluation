```python
# Install libraries (if needed)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Sample ground truth and predictions (simulated)
y_true = np.array([1]*100 + [0]*100)   # 1 = disaster, 0 = non-disaster
y_pred = np.array([1]*82 + [0]*18 + [1]*15 + [0]*85)

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

plt.imshow(cm)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks([0,1], ["Disaster", "Non-Disaster"])
plt.yticks([0,1], ["Disaster", "Non-Disaster"])

for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j], ha='center', va='center')

plt.savefig("confusion_matrix.png")
plt.show()

# Metrics graph
metrics = ["Accuracy", "Precision", "Recall", "F1"]
values = [accuracy*100, precision*100, recall*100, f1*100]

plt.bar(metrics, values)
plt.title("Performance Metrics")
plt.ylabel("Percentage")

plt.savefig("metrics_graph.png")
plt.show()
```
